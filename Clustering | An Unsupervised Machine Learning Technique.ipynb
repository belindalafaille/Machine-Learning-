```python
# Performed k-means clustering to identify new subtypes of a disease to better inform treatment.

#Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#open file
sample_data= scipy.io.loadmat('problemB_data.mat')
# print(sample_data)

data= sample_data['data']
# print(data)
# print(np.shape(data))
```

```python
#- a) Draw the data points as a scatterplot
plt.subplot(2, 2, 1)
plt.scatter(data[:, 0], data[:, 1], marker='x', color='black')
plt.title('Classification of Data: Unknown')

# b) Perform k-means clustering on the data with k= 2 clusters.

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)

plt.subplot(2, 2, 2)
plt.scatter(data[kmeans.labels_ == 0, 0], data[kmeans.labels_ == 0, 1], marker='s', color='red')
plt.scatter(data[kmeans.labels_ == 1, 0], data[kmeans.labels_ == 1, 1], marker='o', color='blue')
plt.title('Clustering: K-means with K=2')

# c) Perform k-means clustering with k=3 clusters

kmeans = KMeans(n_clusters=3, random_state=0).fit(data)

# Plot the clustered data with different markers for each cluster
plt.subplot(2, 2, 3)
plt.scatter(data[kmeans.labels_ == 0, 0], data[kmeans.labels_ == 0, 1], marker='s', color='red')
plt.scatter(data[kmeans.labels_ == 1, 0], data[kmeans.labels_ == 1, 1], marker='o', color='blue')
plt.scatter(data[kmeans.labels_ == 2, 0], data[kmeans.labels_ == 2, 1], marker='^', color='green')
plt.title('Clustering: K-means with K=3')

# Show the plot
plt.tight_layout()
plt.show()
```

Which results seem to closely fit the data? Answer: clusters with K=3 seems to fit the data better.
